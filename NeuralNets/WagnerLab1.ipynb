```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-5,5,100)
y = np.reciprocal(1 + np.exp(-x))
plt.plot(x,y)
plt.title('$y= \dfrac{1}{1+e^{-x}}$')
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```

## Problem 3

```python
from sympy import *
init_printing(use_latex=True)
```

```python
x = symbols('x')
```

```python
y = 1/(1 + exp(-x))
```

```python
y
```

```python
dydx = y.diff(x)
```

```python
dydx
```

```python
x_vals = np.linspace(-5,5,101)
y_vals = np.array([dydx.evalf(subs=dict(x=x_val)) for x_val in x_vals])
plt.plot(x_vals,y_vals)
plt.title('$y=%s$'%(latex(dydx)))
plt.xlabel('x')
plt.xlabel('y')
plt.show()
```
